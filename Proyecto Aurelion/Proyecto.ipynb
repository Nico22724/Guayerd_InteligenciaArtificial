<h1>Estructura</h1>
<h3>Registros de información</h3>

<h4>Clientes</h4>
<ul>
  <li>Id_cliente -- type:numeric <strong>(PK)</strong></li>
  <li>Nombre -- type:text</li>
  <li>Email -- type:text</li>
  <li>Ciudad -- type:text</li>
  <li>Fecha_alta -- type:text</li>
  <li>Relación: FK en <em>Ventas.Id_cliente</em></li>
</ul>

<h4>Detalle_ventas</h4>
<ul>
  <li>Id_venta -- type:numeric <strong>(PK)</strong></li>
  <li>Id_producto -- type:numeric <strong>(FK → Productos.Id_producto)</strong></li>
  <li>Nombre_producto -- type:text</li>
  <li>Cantidad -- type:numeric</li>
  <li>Precio_unitario -- type:numeric</li>
  <li>Importe -- type:numeric</li>
  <li>Relación: FK en <em>Ventas.Id_venta</em></li>
</ul>

<h4>Productos</h4>
<ul>
  <li>Id_producto -- type:numeric <strong>(PK)</strong></li>
  <li>Nombre_producto -- type:text</li>
  <li>Categoria -- type:text</li>
  <li>Precio_unitario -- type:numeric</li>
</ul>

<h4>Ventas</h4>
<ul>
  <li>Id_venta -- type:numeric <strong>(PK)</strong></li>
  <li>Fecha -- type:text</li>
  <li>Id_cliente -- type:numeric <strong>(FK → Clientes.Id_cliente)</strong></li>
  <li>Nombre_cliente -- type:text</li>
  <li>Email -- type:text</li>
  <li>Medio_pago -- type:text</li>
</ul>
